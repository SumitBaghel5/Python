### Q - What is Translator or Compiler ?

Ans - A Translator or Compiler is a software that converts programming language into machine language (0 and 1) so that the computer can understand and execute the program.

Computers only understand binary language (0 and 1).  
The translator helps convert programming language into 0 & 1 language so programs can run on the computer.

**Examples of Translators**
- Compiler
- Interpreter
- Assembler

### Q - What is Python ?
Ans - Python is a programming language developed by Guido van Rossum in 1991.
### -> Features of Python
- Python is a high-level programming language. 

[ ( **High-Level Language** → Easy for humans to read and write.) 

( **Low-Level Language** → It is closer to machine language. It is difficult for humans to read and write but easy for computers to understand and execute.)]
- Python is an interpreted language (executed line by line).
- Python is a case-sensitive language.
- Python has simple and readable syntax.
- Python is open-source and free to use.
- Python supports multiple programming paradigms.
- Python is used in Web Development, Data Science, AI, Machine Learning, and Automation.
